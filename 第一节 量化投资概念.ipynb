### 二、量化投资背景介绍

量化投资是指通过**数量化方式**及**计算机程序化**发出买卖指令，以获取稳定收益为目的的交易方式。

四大特点：

1. 纪律性。纪律性首先表现在依靠模型和相信模型，每一天决策之前，首先要运行模型，根据模型的运行结果进行决策，而不是凭感觉。
2. 系统性。具体表现为“三多”。首先表现在**多层次**，包括在大类资产配置、行业选择、精选个股三个层次上我们都有模型；其次是**多角度**，定量投资的核心投资思想包括宏观周期、市场结构、估值、成长、盈利质量、分析师盈利预测、市场情绪等多个角度；再者就是**多数据**，就是海量数据的处理。
3. 套利思想。通过全面、系统性的扫描捕捉错误定价、错误估值带来的机会。
4. 概率取胜。这表现为两个方面，一是定量投资不断的从历史中挖掘有望在未来重复的历史规律并且加以利用。二是依靠一组股票取胜，而不是一个或几个股票取胜。

投资策略分类：

- 量化选股。
- 量化择时。
- 统计套利。
- 算法交易等。
