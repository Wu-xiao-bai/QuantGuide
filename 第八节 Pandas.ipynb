```python
# 导入pandas模块
import pandas as pd
import numpy as np
```

### 一、Series

类似于array或list。在计算的时候，最好存储同类型的数据。默认会创建自增数字类型的索引。

```python
# 使用列表创建series
_series = pd.Series([1, 2, 3, '东南有大树', 4])  # 使用列表初始化一个Series
_series
```

```python
# 查看其索引
_series.index
```

```python
# 查看所有的数据元素
_series.values
```

```python
# 使用字典创建一个Series
_dict = {"ton": 20, "mary": 18, "东南有大树": 27, "car": None}
_series = pd.Series(_dict)
_series
```

### 二、DataFrame

dataframe相当于一个表格数据。其中的每一行相当于一个series。

```python
# 创建空dataframe
df = pd.DataFrame()
df
```

```python
# 使用list创建dataframe
list_value = [1,2,5,7]
list_index = ['a','b','c','d']
df = pd.DataFrame(list_value, list_index)
df
```

```python
# 采用二维ndarray创建df
_array = np.array([[1,2,5,7], 
                   ['a','b','c','d']])
df = pd.DataFrame(_array)
df
```

```python
# 利用字典数据创建df
_dict = {'value1' : [1,2,5,7], 
         'value2' : ['a','b','c','d']}
df = pd.DataFrame(_dict)
df
```

```python
# 指定列名和索引
_array = np.array([[1,2,5,7], 
                   ['a','b','c','d']])
df = pd.DataFrame(_array, 
                  columns=['值1', '值2', '值3', '值4'], 
                  index=[3, 2])
df
```

### 数据处理

```python
# 如果有缺失值，会被清除掉
df['open'][0] = None
df.dropna().head()
```

```python
# 如果不想去掉缺失值的行，可以对缺失值进行填充
df.fillna(value=0).head()
```

```python
# 判断值是否为空值
pd.isnull(df).head()
```

```python
# 映射，求每一列的平均
df.mean()
```

```python
df.mean()
```

```python
# 使用pandas展示双均线图例
df['ma5'] = df['close'].rolling(5).mean()  # rolling滚动计算
df['ma10'] = df['close'].rolling(10).mean()
df.tail(100)[['ma5', 'ma10']].plot(figsize=(20, 7))
```
